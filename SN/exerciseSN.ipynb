```python
import numpy as np
import matplotlib.pyplot as plt
```

Leggo il file dei dati

```python
data = np.loadtxt("fit_exercise_data.txt")
# Separazione delle colonne: colonna 1 = redshift (z), colonna 2 = modulo della distanza (d)
z = data[:, 0]  # Redshift
mu = data[:, 1]  # Modulo della distanza (magnitudine apparente corretta per SN Ia)
```

plot del grafico 

```python
# Creazione del plot
plt.figure(figsize=(8, 6))
plt.scatter(z, mu, color='b', label="Dati Supernovae", s=10, alpha=0.7)  
plt.xlabel("Redshift (z)")
plt.ylabel("Modulo della distanza mu (mag)")
plt.title("Supernovae Ia: Modulo della distanza vs Redshift")
plt.xscale("log")  # Scala logaritmica per il redshift per una migliore visualizzazione
plt.legend()

#plt.grid(True, which="both", linestyle="--", alpha=0.5)

# Mostra il plot
plt.show()
```

Aggiungiamo gli errori usando la matrice di covarianza fornita dal file

```python
# Caricaro la matrice di covarianza da file 
cov_matrix = np.loadtxt("fit_exercise_covariance.txt")  
```

```python
#calcola errori 
errors = np.sqrt(np.diag(cov_matrix))
```

Grafico con errori

```python
# Creazione del grafico con barre d'errore sulle y
plt.figure(figsize=(8, 5))
plt.errorbar(z, mu, yerr=errors, fmt='o', capsize=5, label="Dati con errori sulle y", color="blue")

# Personalizzazioni del grafico
plt.xlabel("Redshift (z)")
plt.ylabel("Modulo della distanza (mu)")
plt.xscale("log")
plt.title("Dati con errori derivati dalla matrice di covarianza")
plt.legend()
plt.grid(True)
```
